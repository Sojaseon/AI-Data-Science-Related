## Data cleaning

Pandas provides several methods for cleaning data, such as filling missing values and replacing values.


```python
import numpy as np
import pandas as pd

# Create a dataframe with missing values
df = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [4, np.nan, 6],
    'C': [7, 8, 9]
})

print("Dataframe before cleaning:")
print(df)

# Fill missing values with 0
df_filled = df.fillna(0)

print("\nDataframe after filling missing values with 0:")
print(df_filled)
```

## Data manipulation

Pandas provides several methods for manipulating data, such as sorting, filtering, and aggregating.


```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Filter rows where A > 1
df_filtered = df[df['A'] > 1]
print("Dataframe after filtering rows where A > 1:")
print(df_filtered)
```

## Data visualization

Pandas integrates with Matplotlib to provide easy data visualization.

```python
from matplotlib import pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [2, 3, 4, 5, 6],
    'C': [3, 4, 5, 6, 7]
})

# Plot data
df.plot()
plt.show()
```
